## Demographic analysis 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is stored in a DataFrame named 'df'
# If your data is in a different format, adjust the code accordingly

# Load the dataset
# Replace 'your_dataset.csv' with the actual file path or data source


# Demographic Distribution: Age
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Demographic Distribution: Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Distribution of Customer Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Demographic Distribution: Location
plt.figure(figsize=(14, 8))
sns.countplot(x='Location', data=df, palette='viridis', order=df['Location'].value_counts().index)
plt.title('Distribution of Customer Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()

# Identify most common and least common demographic characteristics
most_common_age = df['Age'].mode().values[0]
least_common_age = df['Age'].min()

most_common_gender = df['Gender'].mode().values[0]
least_common_gender = df['Gender'].value_counts().idxmin()

most_common_location = df['Location'].mode().values[0]
least_common_location = df['Location'].value_counts().idxmin()

print(f"Most Common Age: {most_common_age}")
print(f"Least Common Age: {least_common_age}")
print(f"Most Common Gender: {most_common_gender}")
print(f"Least Common Gender: {least_common_gender}")
print(f"Most Common Location: {most_common_location}")
print(f"Least Common Location: {least_common_location}")
```

## Customer Loyalty Analysis:

Explore factors contributing to customer loyalty, such as previous purchases or subscription status.
Identify any trends or patterns in repeat purchases.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is stored in a DataFrame named 'df'
# If your data is in a different format, adjust the code accordingly

# Load the dataset
# Replace 'your_dataset.csv' with the actual file path or data source


# Check for missing values in 'Previous Purchases' and 'Frequency of Purchases'
missing_values = df[['Previous Purchases', 'Frequency of Purchases']].isnull().sum()
print("Missing Values:\n", missing_values)

# Drop rows with missing values in 'Previous Purchases' and 'Frequency of Purchases'
df_cleaned = df.dropna(subset=['Previous Purchases', 'Frequency of Purchases'])

# Explore factors contributing to customer loyalty
plt.figure(figsize=(12, 8))

# Subplot 1: Explore the impact of previous purchases on loyalty
plt.subplot(2, 1, 1)
sns.boxplot(x='Subscription Status', y='Previous Purchases', data=df_cleaned, palette='viridis')
plt.title('Impact of Subscription Status on Previous Purchases')
plt.xlabel('Subscription Status')
plt.ylabel('Previous Purchases')

# Subplot 2: Explore the impact of subscription status on loyalty
plt.subplot(2, 1, 2)
sns.countplot(x='Subscription Status', hue='Frequency of Purchases', data=df_cleaned, palette='Set2')
plt.title('Impact of Subscription Status on Purchase Frequency')
plt.xlabel('Subscription Status')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
